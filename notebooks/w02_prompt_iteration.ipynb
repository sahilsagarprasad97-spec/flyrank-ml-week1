# Week 2 - Prompt Iteration

## Objective

The goal of this assignment is to improve a weak AI prompt step by step by changing only one thing at a time and comparing the results.

## Version 0 (Baseline)

### Prompt

### Sample Output

The dataset contains information that can be analyzed to understand patterns and relationships between variables.

### Notes

**What changed?**
Nothing. This is the original weak prompt.

**What improved?**
Nothing yet.

**What still failed?**
The response is too generic and lacks context.

**Next Step**
Add a clear goal.

# Reflection

Improving the prompt one step at a time helped me understand how each change affected the AI's response. Small improvements such as defining the goal, audience, output format, quality criteria, and verification made the final output much clearer and more useful.
