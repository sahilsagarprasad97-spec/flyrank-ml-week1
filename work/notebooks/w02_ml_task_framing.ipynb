Purchased

The target column is **Purchased**, where:
- 1 = Customer purchased the product
- 0 = Customer did not purchase the product

Customer behavior depends on many factors such as age, income, and website visits. Machine Learning can learn these patterns automatically, while fixed rules cannot adapt well to new data.
